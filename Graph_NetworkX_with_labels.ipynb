#Random Walk


*   Traversing the graph through the connected nodes
*   https://www.analyticsvidhya.com/blog/2019/11/graph-feature-extraction-deepwalk/



```python
import random
def get_randomwalk(graph_val, node, path_length):

  random_walk = [node]

  for i in range(path_length-1):
    temp = list(graph_val.neighbors(node))
    temp = list(set(temp) - set(random_walk))
    if len(temp) == 0:
      break

    random_node = random.choice(temp)
    random_walk.append(random_node)
    node = random_node

  return random_walk
```

```python
get_randomwalk(G, 'E', 5)
```

# The Total number of walks

```python
from tqdm import tqdm
# get list of all nodes from the graph
all_nodes = list(G.nodes())

random_walks = []
for n in tqdm(all_nodes):
  for i in range(1):
    random_walks.append(get_randomwalk(G, n, 5))

# count of sequences
len(random_walks)
```
